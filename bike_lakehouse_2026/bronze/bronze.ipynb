# Reading from csv files and writing to Bronze Layer

```python

base_path = "/Volumes/workspace/bronze/source_systems/"
source_table_mapping = {
    "source_crm/sales_details.csv": "crm_sales_details",
    "source_crm/cust_info.csv": "crm_cust_info",
    "source_crm/prd_info.csv": "crm_prd_info",
    "source_erp/CUST_AZ12.csv": "erp_cust_az12",
    "source_erp/LOC_A101.csv": "erp_loc_a101",
    "source_erp/PX_CAT_G1V2.csv": "erp_px_cat_g1v2",
}

for source, table in source_table_mapping.items():
    df = (
        spark.read.option("header", "true")
        .option("inferSchema", "true")
        .csv(f"{base_path}/{source}")
    )
    df.write.mode("overwrite").saveAsTable(f"workspace.bronze.{table}")
```
